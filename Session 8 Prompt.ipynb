```python
from matplotlib import pyplot as plt
from numpy import *

x = linspace(0,1,100)

def sine(n):
    return sin(n)
def cosine(n):
    return cos(n)
```

```python
fig,(ax1,ax2) = plt.subplots(1,2)
fig.suptitle('Sine x Cosine')
ax1.plot(x,sine(x))
ax2.plot(x,cosine(x))
ax1.set_title('sine(x)')
ax2.set_title('cos(x)')
plt.savefig("Sin(x) x Cos(x).pdf", format="pdf", bbox_inches="tight")
```
